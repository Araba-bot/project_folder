## cleaning up the data

```python
#fill in missing profits by creating a new column for it that calculates all the profits
for ind, row in car_sales.iterrows():
    car_sales.loc[ind, "sale_profit"] = row['sale_price'] - row['purchase_price']
```

```python
car_sales.head(20)
```

```python
## replace missing year number with unknown
car_sales["year"].fillna("unknown", inplace=True)
car_sales.head(20)
```

```python
## replace missing color with missing
car_sales["color"].fillna("Missing", inplace=True)
car_sales.tail(20)
```

```python
## drop missing data with car_model and car make
car_sales.dropna(subset=['car_model', 'car_make'], inplace=True)
```
